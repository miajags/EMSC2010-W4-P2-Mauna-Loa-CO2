This record is referred to as the *Keeling curve* in honour of Charles Keeling who started the monitoring program and ran it until his death in 2005.

```python
spreadsheet_name = 'CO2 Data.xlsx'
worksheet_name = 'weekly_data'
df = pd.read_excel(spreadsheet_name)
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 4)) #make axes (called ax) 10 units wide and 4 units high
ax.plot(df['Date'],df['CO2'],'k') #plot time vs. Z on ax as a black ('k') line
ax.set_xlabel('Date') #text for the x-axis label
ax.set_ylabel('CO2') #text for the y-axis label
ax.set_xlim([1957-1-1,2024-12-30]) #interval for the x-axis
ax.set_ylim([300,440]) #interval for the y-axis
ax.minorticks_on() #show the minor ticks
#run this, focus on handling data not writing code
```

```python
spreadsheet_name = 'co2_mm_mlo.xlsx'
worksheet_name = 'co2_mm_mlo'
df = pd.read_excel(spreadsheet_name, sheet_name=worksheet_name)
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 4)) #make axes (called ax) 10 units wide and 4 units high
ax.plot(df['decimal date'],df['average'],'b') #plot time vs. Z on ax as a black ('k') line
ax.set_xlabel('Date') #text for the x-axis label
ax.set_ylabel('CO2 average (ppm)') #text for the y-axis label
ax.set_xlim([1957,2026]) #interval for the x-axis
ax.set_ylim([310,440]) #interval for the y-axis
ax.minorticks_on() #show the minor ticks
ax.set_title('Mauna Loa Monthly CO2')
#run this, focus on handling data not writing code
```
